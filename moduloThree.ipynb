<h1>Condiciones y ramificaciones</h1>

<p>x == y --> x es igual a y?</p>
<p>x > y --> x es mayor a y?</p>
<p>x < y --> x es menor a y?</p>
<p>x >= y --> x es mayor o igual a y?</p>
<p>x <= y --> x es menor o igual a y?</p>
<p>x != y --> x es distinto a y?</p>

```python
a = 6

print(a == 7)
print(a == 6)
print(a > 3)
print(a <= 6)
print(a != 6)
```

```python
edad = 17
padres = True
#Si la comparacion es verdadera.
if (edad >= 18):
    print("Eres mayor de edad, puedes ingresar al concierto.") #Solo si eres mayor de edad imprime esto.
elif (padres == True): #Si eres menor de edad pero vas con alguien mayor imprime esto:
    print("Eres menor de edad pero, entras con un adulto mayor al concierto.")
else: #Si la comparacion es falsa.
    print("Eres menor de edad, no puedes ingresar al concierto.")#Si no lo eres imprime esto.
```

# Operadores Logicos

not(False) --> True

not(True) --> False

or --> Si a o b es mayor que c. a > c or b > c

and --> Si a y b son mayor que c. a > c and b > c

% --> Residuo

```python
print(not(True))
a = 1
b = 3
c = 2
print("\n")
#Compara si uno de los dos se cumple
print(a > c or b > c)
print("\n")
#Compara si ambos se cumplen
print(a > c and b > c)
print("\n")
print(10 % 2) #10 divido 2 deja un residuo de 0
```
